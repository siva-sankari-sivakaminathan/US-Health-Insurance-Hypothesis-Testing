```python
#Identify potential outliers

numerical_cols = ['age', 'bmi','children', 'charges']  # Adjust based on your data
for col in numerical_cols:
    if df[col].dtype != 'object':  # Check for non-numerical columns
        z = np.abs(stats.zscore(df[col]))
        # Threshold for outliers (adjust as needed, common range: 3-4)
        outliers_z = df[col][z > 3]
        print(f"Potential outliers in {col} using Z-scores:")
        print(outliers_z.describe())  # Descriptive statistics of outliers
```

```python
def iqr_outliers(col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[col] < lower_bound) | (df[col] > upper_bound)]

outliers_iqr = iqr_outliers('charges')  # Example for 'charges' column
print(f"Potential outliers in charges using IQR:")
print(outliers_iqr.describe())  # Descriptive statistics of outliers
```

```python
def winsorize(col, lower_bound, upper_bound):
 
  df.loc[df[col] < lower_bound, col] = lower_bound  # Cap values below lower bound
  df.loc[df[col] > upper_bound, col] = upper_bound  # Cap values above upper bound
  return df
```

```python
# Calculate IQR for charges
q1 = df['charges'].quantile(0.25)
q3 = df['charges'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr  # Lower winsorization bound (adjust multiplier as needed)
upper_bound = q3 + 1.5 * iqr  # Upper winsorization bound (adjust multiplier as needed)

# Winsorize the charges column
df_winsorized = winsorize('charges', lower_bound, upper_bound)
```

**HYPOTHESIS TESTING**

**Impact of Age**

o	H0: There is no significant difference in average healthcare charges between younger and older adults.

o	H1: Average healthcare charges increase with age.


```python
import statsmodels.api as sm 

n_unique_ages = df['age'].nunique()

#t-test 

if n_unique_ages == 2:
    younger_group = df[df['age'] == df['age'].min()]['charges']
    older_group = df[df['age'] == df['age'].max()]['charges']
    tstat, pval = stats.ttest_ind(younger_group, older_group)
    print(f"t-test: t={tstat:.4f}, p-value={pval:.4f}")
else:
    print("t-test not applicable: More than 2 age groups found.")
```

```python
if n_unique_ages >= 3:
      model = sm.formula.ols('charges ~ age', data=df).fit()
      anova_table = sm.stats.anova_lm(model)
      print(anova_table)
else:
      print("ANOVA not applicable: Less than 3 age groups found.")
```

Sum of squares indicates a large amount of variation in charges explained by the age factor.

A very high F-statistic, suggesting a strong potential difference in charges between groups.

Extremely small p-value, indicating very strong evidence against the null hypothesis (all age groups have equal mean charges). There's likely a significant difference in charges based on age.

**Gender and Charges**

o	H0: There is no significant difference in average healthcare charges between men and women.

o	H1: Men or women have higher average charges.


```python
# Filter data by gender
male_charges = df[df['sex'] == 'male']['charges']
female_charges = df[df['sex'] == 'female']['charges']

# Perform t-test (assuming equal variances)
tstat, pval = stats.ttest_ind(male_charges, female_charges, equal_var=True)

# Print the results
print(f"t-test: t={tstat:.4f}, p-value={pval:.4f}")
```

There's not enough evidence to reject the null hypothesis at a common significance level of 0.05 (or 5%). In other words, based on the data you analyzed, we cannot statistically conclude that the average insurance charges differ between males and females.

**BMI and Charges**

o	H0: There is no significant correlation between BMI and healthcare charges.

o	H1: Higher BMI is associated with higher charges.


```python
# Calculate Pearson's correlation coefficient
correlation, p_value = stats.pearsonr(df['charges'], df['bmi'])

# Print results
print(f"Pearson's correlation coefficient: {correlation:.4f}")
print(f"p-value: {p_value:.4f}")
```

This indicates a weak correlation between bmi and charges.

There's a statistically significant, but weak, negative correlation between the variables. This means that there might be a slight tendency for one variable to decrease as the other increases, but the strength of this association is weak. The coefficient value of -0.0730 suggests a very minimal negative effect.

**Smoking and Charges**

o	H0: Smoking status has no significant impact on healthcare charges.

o	H1: Smokers have higher average charges.


```python
# Filter data by smoking status
smoker_charges = df[df['smoker'] == 'yes']['charges']
non_smoker_charges = df[df['smoker'] == 'no']['charges']

# Perform t-test (assuming equal variances)
tstat, pval = stats.ttest_ind(smoker_charges, non_smoker_charges, equal_var=True)

# Print the results
print(f"t-test: t={tstat:.4f}, p-value={pval:.4f}")
```

Based on the t-test results, we can conclude that there's a statistically significant difference in insurance charges between smokers and non-smokers. The positive t-statistic suggests that, on average, smokers likely have higher insurance charges compared to non-smokers.

**Region and Charges**

o	H0: There is no significant difference in average healthcare charges across different regions.

o	H1: Healthcare charges vary significantly between regions.


```python
#check for normality

# Group data by region
grouped_by_region = df.groupby('region')

# Visual Inspection (Histogram and Q-Q plot for each region)
for region, group_data in grouped_by_region:
    plt.figure(figsize=(10, 4))
    
    # Histogram
    plt.subplot(1, 2, 1)
    group_data['charges'].hist(bins=20, edgecolor='black')
    plt.title(f"Charges Distribution (Region: {region})")
    plt.xlabel("Charges")
    plt.ylabel("Frequency")

   

```

```python
# Statistical Tests (Shapiro-Wilk test for each region)
for region, group_data in grouped_by_region:
    shapiro_wilk_stat, pval = stats.shapiro(group_data['charges'])
    print(f"Region: {region} - Shapiro-Wilk Test Statistic: {shapiro_wilk_stat:.4f}, p-value: {pval:.4f}")
```

Since all regions have p-values much lower than the common significance level of 0.05, it's safe to conclude that the "charges" variable is not normally distributed across all regions.

This finding indicates that using ANOVA to compare charges across regions might not be appropriate. ANOVA relies on the assumption of normality of residuals, which is likely violated here.
A more suitable test for comparing charges in this scenario would be the Kruskal-Wallis test, a non-parametric test that does not require normality assumptions.

```python
def check_and_fix_charges(data_by_region):
  # Get the expected length of the charges list based on any existing region
  expected_len = len(next(iter(data_by_region.values())))  # Get length from any region's charges
 # Loop through regions and fix charges if needed
  for region, charges in data_by_region.items():
    if len(charges) != expected_len:
        data_by_region[region] = charges + [0] * (expected_len - len(charges))  # Pad with zeros
        return data_by_region

# Group data by region
grouped_by_region = df.groupby('region')

# Extract charges for each region (assuming 'charges' is the column name)
charges_by_region = {region: group_data['charges'].tolist() for region, group_data in grouped_by_region}

# Fix charges if necessary to ensure consistent shape
charges_by_region = check_and_fix_charges(charges_by_region.copy())  # Pass a copy to avoid modifying original dict

# Perform Kruskal-Wallis test
hstat, pval = stats.kruskal(*charges_by_region.values())
print(f"Kruskal-Wallis test: H={hstat:.4f}, p-value={pval:.4f}")

# Optional: Analyze results (e.g., print interpretation based on p-value)
if pval < 0.05:
  print("Reject null hypothesis: There's a statistically significant difference in charges across regions.")
else:
  print("Fail to reject null hypothesis: No significant difference detected.")
```

**Children and Charges**

o	H0: The number of children has no significant effect on healthcare charges.

o	H1: More children lead to higher charges.


```python
model = sm.formula.ols('charges ~ children', data=df).fit()
anova_table = sm.stats.anova_lm(model)
print(anova_table)
```

This very low p-value (less than 0.05, commonly used as a significance threshold) indicates that the observed difference in charges based on the number of children is unlikely to be due to chance. In other words, there's statistically significant evidence to suggest that the number of children has an impact on insurance charges.

The analysis demonstrates that the number of children can be a statistically significant factor influencing insurance charges in some regions. Further investigation, such as post-hoc tests, might be necessary to identify specific groups with statistically different charges. It's also important to consider other factors that could influence charges alongside the number of children.
